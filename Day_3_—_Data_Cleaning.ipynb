# **3. Handle Missing Values**

```python
# Drop missing rows
df_drop = df.dropna()

# Fill numeric column
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill categorical column â€” only if exists
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

# Replace 0 salary with NaN
if 'Salary' in df.columns:
    df['Salary'] = df['Salary'].replace(0, np.nan)

df
```

# **4. Handle Duplicates**

```python
df.drop_duplicates(inplace=True)
df
```

# **5. Handle Outliers**

Using IQR (Interquartile Range) method:

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Filter
df_clean = df[(df['Age'] >= lower) & (df['Age'] <= upper)]
df
```

# **6. Handle Incorrect Data Types**

```python
# handle incorrect data types

df['Age'] = df['Age'].astype(int)
df
```

# **7. Feature Standardization (Optional)**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
df
```

# **8. Export Clean Data**

```python
df.to_csv('cleaned_data.csv', index=False)
```
