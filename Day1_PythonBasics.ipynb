# **Core Python Concepts to Master**

```python
# Variables and Data Types

a = 10
b = 3.14
c = "Abhi"
d = True
print(type(a), type(b), type(c), type(d))
```

```python
# Lists, Tuples, Sets, Dictionaries
# Lists are mutable, Tuples are immutable, Sets remove duplicates, and Dictionaries store key-value pairs.

my_list = [1, 2, 3, 4]
my_tuple = (10, 20, 30)
my_set = {1, 2, 2, 3}
my_dict = {'name': 'Abhi', 'age': 25}

print(my_list, my_tuple, my_set, my_dict)
```

# **Conditional & Looping**

```python
# If-Else
x = 5
if x > 0:
    print("Positive")
else:
    print("Negative")

# For loop
for i in range(1, 6):
    print(i)
```

# **Functions**

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Abhi"))
print(greet("World!"))
```
